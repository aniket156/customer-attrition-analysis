**DATA PRE-PROCESSING**

*   Unnecessary features would decrease the training speed, the model interpretability and the generalization performance on the test data.
*  Therefore, finding and selecting the most useful features in the dataset is crucial.
*  Assigning input features to X and output (Churn) to y

```python
x = df.drop(["class", "area_code", "phone_number"], axis = "columns") # area_code and phone_number features are irrelevant to proceed further to train the model
y = df["class"]
```

**DATA VISUALIZATION**

```python
df.hist(figsize = (30, 30))
plt.show()
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 150)
```

**Function for finding the accuracy**

```python
def metrics(h, y):
  h = np.array(h).reshape(-1)
  y = np.array(y).reshape(-1)

  tp = sum((h==y) & (y==1))
  tn = sum((h==y) & (y==0))
  fp = sum((h!=y) & (y==0))
  fn = sum((h!=y) & (y==1))

  # print(tp, tn, fp, fn)

  accuracy = (tp+tn) / (tp+tn+fp+fn)

  d = {'accuracy':accuracy}

  return d
```

**Building the NEURAL NETWORK MODEL**

```python
import tensorflow as tf
net = tf.keras.Sequential([
        tf.keras.layers.Dense(10, input_shape=(x_train.shape[1], ), activation='tanh'), 
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dense(3, activation='tanh'), 
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
net.compile(optimizer='adam', 
              loss = 'binary_crossentropy', 
              metrics=['accuracy'])
```

```python
net.fit(x_train, y_train, epochs=25)
```

```python
y_predictNN = net.predict(x_test)
```

```python
print(y_predictNN[:10])
```

```python
predictions = net.predict(x_test.iloc[:7]).reshape(-1)
predictions[predictions> 0.40] = 1
predictions[predictions<=0.40] = 0

df = x_test.iloc[:7].copy()
df['predictions'] = predictions
df
```

**Finding the Accuracy of the NN-model**

```python
metrics(net.predict(x_test)>0.5, y_test)
```
